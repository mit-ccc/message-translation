## Translation

Next, we perform a full translation for all of the shared words

```python
a_words = set(forward.w2idA.keys())
b_words = set(forward.id2wB.values())
shared = set.intersection(a_words, b_words)
shared = list(shared)
```

```python
translations = forward.translate_words(shared, k=5)
```

```python
translations_tuples = [(w, s) for w, s in zip(translations[0], translations[1])]
```

```python
for word, neighbors in zip(shared, translations[0]):
    print(word, neighbors)
```

```python
correct_count = 0
total_count = len(shared)
for i in range(len(shared)):
    nn_neighbors = translations_tuples[i][0]
    word = shared[i]
#     print(nn_neighbors, word)
    if word in set(nn_neighbors):
        correct_count += 1

acc = correct_count / total_count
print("Accuracy@5 is ", acc)
```
